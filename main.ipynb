```python
import pandas as pd

df = pd.read_csv('Customer_Data.csv')
df.head()
```

Calculating Gender Count and Percentage

```python
gender_count = df['Gender'].value_counts()
gender_percentage = (gender_count / gender_count.sum()) * 100

result = pd.DataFrame({
    'Gender_Count': gender_count,
    'Gender_Percentage': gender_percentage})

result
```

Calculating Contract Count and Percentage


```python
contract_count = df['Contract'].value_counts()
contract_percentage = (contract_count / contract_count.sum()) * 100

result = pd.DataFrame({
    'Contract_Count' : contract_count,
    'Contract_Percentage' : contract_percentage})
result
```

Calculating Customers status Count and Percentage


```python
customer_status_count = df['Customer_Status'].value_counts()
customer_status_percentage = (customer_status_count / customer_status_count.sum()) * 100

result = pd.DataFrame({
    'Customer_Status_Count' : customer_status_count,
    'Customer_Status_Percentage' : customer_status_percentage})
result
```
